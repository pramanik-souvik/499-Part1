```python
import pandas as pd

# Load your dataset
df = pd.read_csv("final_data.csv")  # Replace with your file name

# Drop rows where 'content', 'context', and 'title' are all the same
df_filtered = df[~((df['content'] == df['context']) & (df['context'] == df['title']))]

# Optionally reset index (optional)
df_filtered = df_filtered.reset_index(drop=True)

# Save to new file (or overwrite original)
df_filtered.to_csv("filtered_dataset.csv", index=False)

print(f"Removed {len(df) - len(df_filtered)} rows with identical content, context, and title.")
```
